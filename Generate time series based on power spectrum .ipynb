```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 5
wg = 10
kesi = 0.25

def Sx(w,alpha,wg,kesi):
    return (1 + alpha*w*w)/((wg*wg - w*w)**2 + (2*kesi*w*wg)**2)
```

```python
Ultimate_frequency = 100

w_sample = np.linspace(0,Ultimate_frequency,1000)
Sx_sample = Sx(w,alpha,wg,kesi)

plt.plot(w_sample,Sx_sample)
plt.show()
```

```python
print(4*Sx(0.1*1,alpha,wg,kesi)*0.1*np.sin(0.1*1*1+np.random.uniform(0,(np.pi)*2)))
```

```python
frequency_space = 0.1
N = Ultimate_frequency/frequency_space

```

```python
Ultimate_time = 100
time_space = 0.1
N1 = int(Ultimate_time/time_space)

X_t = np.zeros((N1,))


i = 0
while i<N1:
    j = 0
    while j<N:
        inter = np.abs(4*Sx(frequency_space*j,alpha,wg,kesi)*frequency_space*np.sin(frequency_space*j*i+np.random.uniform(0,(np.pi)*2)))
        X_t[i] += inter**0.5
        j = j+1
    i = i+1

```

```python
time = np.linspace(0,Ultimate_time,N1)
plt.plot(time,X_t)
plt.show()
```
